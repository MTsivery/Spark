# 5- Spark MLLib

```python
#Import necessary library
from pyspark.sql import SparkSession
from pyspark.ml.regression import LinearRegression
from pyspark.mllib.linalg import Vectors
spark = SparkSession.builder.appName('MLLIB Usage Example').getOrCreate()
#Utilisons le dataset diamonds.csv
dataset = spark.read.format('com.databricks.spark.csv').options(header='true', inferschema='true').load('/content/diamonds.csv')
```

```python
dataset = dataset.withColumnRenamed('price','label')
```

```python
#Code here
dataset = dataset['label','carat','x','y','z']
```

```python
#Transform the features into vectors (carat,depth,x,y,z)
from pyspark.ml.feature import VectorAssembler

ignore = ['label']
assembler = VectorAssembler(
    inputCols=[x for x in dataset.columns if x not in ignore],
    outputCol='features')

training= assembler.transform(dataset) ['label', 'features']
```

```python
training.show(n=2)
```

```python
training.dtypes
```

```python
# These are the default values for the featuresCol, labelCol, predictionCol
lr = LinearRegression(featuresCol='features', labelCol='label', predictionCol='prediction')
```

```python
# Fit the model
lrModel = lr.fit(training)
```

```python
# Print the coefficients and intercept for linear regression
print("Coefficients: {}".format(str(lrModel.coefficients))) # For each feature...
print('\n')
print("Intercept:{}".format(str(lrModel.intercept)))
```

```python
# Summarize the model over the training set and print out some metrics
trainingSummary = lrModel.summary
```

```python
trainingSummary.residuals.show()
print("RMSE: {}".format(trainingSummary.rootMeanSquaredError))
print("r2: {}".format(trainingSummary.r2))
```

```python
from pyspark import SparkContext
from pyspark.streaming.context import StreamingContext
from pyspark import SparkConf
sparkConf = SparkConf()
sparkConf.setAppName("Text Cleaning")
sc = SparkContext.getOrCreate(sparkConf)
strc = StreamingContext(sc, 3)


text_data = strc.socketTextStream("localhost", 8083)
```

```python
import re
import nltk
nltk.download()
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(sentence):
    sentence = sentence.lower()
    sentence = re.sub("\s+"," ", sentence)
    sentence = re.sub("\W"," ", sentence)
    sentence = re.sub(r"http\S+", "", sentence)
    sentence = ' '.join(word for word in sentence.split() if word not in stop_words)
    sentence = [lemmatizer.lemmatize(token, "v") for token in sentence.split()]
    sentence = " ".join(sentence)
    return sentence.strip()
```

```python
cleaned_text = text_data.map(lambda line: clean_text(line))
cleaned_text.pprint()
```

Pour lancer notre traitement, contrairement à Spark où l’exécution d’une action finale lançait un traitement, Spark Streaming nécessite d’appeler la méthode sc.start().

Le traitement démarre dans des threads séparés. Il faut donc empêcher le thread principal de s’arrêter. Le plus simple est d’appeler sc.awaitTermination() mais il est possible d’attendre un évènement particulier ou d’exécuter un autre traitement.

```python
strc.start()
strc.awaitTermination()
```
